#### Define

Create a new column that shows us the best dog breed, to help us predict

#### Code

```python
def predict_breed(row):
    if row.p1_dog:
        return row.p1
    elif row.p2_dog:
        return row.p2
    elif row.p3_dog:
        return row.p3
df_2_clean22['dog_breed'] = df_2_clean22.apply(lambda row: predict_breed(row),axis=1)
```
